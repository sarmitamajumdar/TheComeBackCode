# Human Vs AI

```python
import matplotlib.pyplot as plt
import numpy as np

# Years and job data (in millions)
years = np.array(range(2020, 2031))
human_jobs = np.array([100, 97, 93, 88, 82, 75, 65, 55, 42, 30, 20])
ai_jobs = np.array([5, 8, 13, 20, 30, 42, 55, 70, 85, 95, 110])

# Create mask where AI jobs exceed human jobs
ai_surpasses_humans = ai_jobs > human_jobs

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, human_jobs, label='Human Jobs', color='blue', linewidth=2, marker='o')
plt.plot(years, ai_jobs, label='AI Jobs', color='red', linewidth=2, marker='s')

# Highlight crossover zone
plt.fill_between(years, human_jobs, ai_jobs, where=ai_surpasses_humans,
                 interpolate=True, color='lightcoral', alpha=0.3,
                 label='AI Dominance Zone')

# Formatting
plt.title('Human vs AI Jobs (20202030)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Jobs (in millions)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# Routine-vs-purpose

```python
import matplotlib.pyplot as plt
import numpy as np

# Define data as NumPy arrays for element-wise comparison
years = np.array(range(2020, 2031))
routine_tasks = np.array([90, 85, 78, 70, 60, 50, 42, 35, 28, 20, 15])
purpose_roles = np.array([10, 15, 22, 30, 40, 50, 58, 65, 72, 80, 85])

# Create Boolean mask for fill_between
mask = purpose_roles > routine_tasks

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, routine_tasks, label='Routine Tasks', color='red', marker='o', linewidth=2)
plt.plot(years, purpose_roles, label='Purpose-Driven Roles', color='green', marker='s', linewidth=2)

# Highlight crossover zone
plt.fill_between(years, purpose_roles, routine_tasks, where=mask,
                 interpolate=True, color='lightgreen', alpha=0.3,
                 label='Human Purpose Zone')

# Customize plot
plt.title('Shift from Routine to Purpose (20202030)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Role Intensity (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
